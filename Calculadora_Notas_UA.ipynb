# **Calculadora de Notas**

Con este programa podes calcular cuantos puntos necesitas en tu examen final para pasar la materia

En caso de que no puedas exonerar, ingresa tus dos mejores parciales para calcular tus notas posibles:

```python
p1=int(input('Igresa tu nota del Parcial 1 (Escala de 0 a 100): '))
p2=int(input('Igresa tu nota del Parcial 2 (Escala de 0 a 100): '))

#los parciales equivalen al 60% de la nota final
#el final equivale al 40% de la nota final

suma_parciales= (p1*0.30+p2*0.30)

print(f'\nTu sumatoria de parciales es de: {suma_parciales}')

print('------------------------------------------------------------------')
print('\nLista de notas posibles:' )
print(' ')
print(f'Para pasar con 2, Debes hacer: {(60-suma_parciales)/0.4}')
print(' ')
print(f'Para pasar con 3, Debes hacer: {(70-suma_parciales)/0.4}')
print(' ')
print(f'Para pasar con 4, Debes hacer: {(80-suma_parciales)/0.4}')
print(' ')
print(f'Para pasar con 5, Debes hacer: {(91-suma_parciales)/0.4}')
print(' ')

```
